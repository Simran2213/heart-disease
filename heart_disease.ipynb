# spliting the features(imp colmns to consider) and target(result)

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x = data.drop(columns='target',axis=1)
y = data['target']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x_train
```

```python
x_test
```

```python
x.shape,x_train.shape,x_test.shape
```

# model training
#logistic regression

```python
model = LogisticRegression()
```

# traing our mpodel

```python
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

# model evaluation             accuracy score

```python
#accuracy on training data
```

```python
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
```

```python
print('accuracy on training data:',training_data_accuracy)
```

```python
x_test_prediction = model.predict(x_test)
testing_data_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python
print('accuracy on testing data:',testing_data_accuracy)
```

# predicting system

```python
input_data = (67,1,0,160,286,0,0,108,1,1.5,1,3,2)
#change input data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshaping the numpy array as we are predicting for one data set instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1) 
prediction = model.predict(input_data_reshaped)
prediction
```

```python
if (prediction[0]==0):
    print('PATIENT DO NOT HAVE HEART DISEASE')
else:
    print('PATIENT HAVE HEART DISEASE')
```
